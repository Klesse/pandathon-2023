```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/final.csv')
```

```python
df
```

```python
group_sizes = df.groupby('UF').size()
group_sizes.sort_values(inplace=True, ascending = False)
# Alternatively, you can use count() for non-null values
# group_sizes = df.groupby('Category').count()['Value']

group_sizes.plot(kind='bar')
plt.title('Secretarios por Estado')
plt.xlabel('Estados')
plt.ylabel('N Secretarios')
plt.show()
```

```python
funcgroup = df.groupby('Funcao').size()
funcgroup.sort_values(inplace=True, ascending = False)
plotfunc=funcgroup.head(5)
plotfunc.plot(kind='bar', color='green')
plt.title('Secretarios por Funcionalidade')
plt.xlabel('Funcionalidades')
plt.ylabel('N Secretarios')
plt.show()
```

```python
citgp = df.groupby('Cidade').size()
citgp.sort_values(inplace=True, ascending = False)
# Alternatively, you can use count() for non-null values
# group_sizes = df.groupby('Category').count()['Value']
finalcit=citgp.head(10)
finalcit.plot(kind='bar', color='red')
plt.title('Secretarios por Cidade')
plt.xlabel('Cidades')
plt.ylabel('N Secretarios')
plt.show()
```
